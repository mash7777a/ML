```python
def check(x):
  if (x['fall'] == 'Statistical Learning Theory') | (x['fall']  == 'Высокопроизводительные вычисления'):
    val = 60
  elif x['fall']  == 'Анализ неструктурированных данных':
      val = 400
  else:
    val = 30
  return val
```

```python
# количество мест на курсе
uniques = pd.DataFrame({"fall": pd.unique(data[['fall_1',	'fall_2',	'fall_3']]. values.ravel ())})
uniques['places'] = uniques.apply(check, axis=1)
```

```python
data = data.sort_values(by=['percentile'])
df = pd.DataFrame(index=data['id'])
df['course1'] = ''
df['course2'] = ''
```

```python
data['is_first_place'] = NaN
data['is_last_place'] = NaN
```

```python
percents = pd.DataFrame(index=uniques['fall'])
percents['best'] = 2
percents['best_id'] = ''
percents['worst'] = -1
percents['worst_id'] = ''
```

```python
def wave1(id, perc, f_1, f_2, n_c):
  ind1 = int(uniques.index[uniques['fall'] == f_1].tolist()[0])
  ind2 = int(uniques.index[uniques['fall'] == f_2].tolist()[0])

  if uniques.at[ind1, 'places'] > 0:

    if perc < percents.at[f_1, 'best']:
      percents.at[f_1, 'best'] = perc
      percents.at[f_1, 'best_id'] = id
    if perc > percents.at[f_1, 'worst']:
      percents.at[f_1, 'worst'] = perc
      percents.at[f_1, 'worst_id'] = id

    df.at[id, 'course1'] = f_1
    uniques.at[ind1, 'places'] -= 1

  if uniques.at[ind1, 'places'] == 0:
    df.at[id, 'course1'] = '???'

  df.at[id, 'course2'] = '-'
  if  n_c == 2:

    if (f_1 == f_2) | (uniques.at[ind2, 'places'] == 0):
        df.at[id, 'course2'] = '???'

    elif uniques.at[ind2, 'places'] > 0:

      if perc < percents.at[f_2, 'best']:
        percents.at[f_2, 'best'] = perc
        percents.at[f_2, 'best_id'] = id
      if perc > percents.at[f_2, 'worst']:
        percents.at[f_2, 'worst'] = perc
        percents.at[f_2, 'worst_id'] = id

      df.at[id, 'course2'] = f_2
      uniques.at[ind2, 'places'] -= 1
  return 0
```

```python
data.apply(lambda x: wave1(x['id'], x['percentile'], x['fall_1'], x['fall_2'], x['number_courses']), axis=1)
```

```python
def sorting(b_i, w_i):
  ind1 = int(data.index[data['id'] == b_i].tolist()[0])
  data.at[ind1, 'is_first_place'] = True
  ind2 = int(data.index[data['id'] == w_i].tolist()[0])
  data.at[ind2, 'is_last_place'] = True
  return 0
percents.apply(lambda x: sorting(x['best_id'], x['worst_id']), axis=1)
```

```python
def wave2(id, f_2, f_3, c_1, c_2):
  ind1 = int(uniques.index[uniques['fall'] == f_2].tolist()[0])
  ind2 = int(uniques.index[uniques['fall'] == f_3].tolist()[0])
  ind_id = int(data_wave.index[data_wave['id'] == id].tolist()[0])

  if c_2 != '-':
    if c_1 == '???':
      if uniques.at[ind2, 'places'] > 0:
        df.at[id, 'course1'] = f_3
        data_wave.at[ind_id, 'course1'] = f_3
        uniques.at[ind2, 'places'] -= 1
    elif c_2 == '???':
      if uniques.at[ind2, 'places'] > 0:
        df.at[id, 'course2'] = f_3
        data_wave.at[ind_id, 'course2'] = f_3
        uniques.at[ind2, 'places'] -= 1

  else:
    if uniques.at[ind1, 'places'] > 0:
      df.at[id, 'course1'] = f_2
      data_wave.at[ind_id, 'course1'] = f_2
      uniques.at[ind1, 'places'] -= 1
  return 0
```
